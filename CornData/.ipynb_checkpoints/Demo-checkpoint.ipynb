#### make_bins() funct

```python
def make_bins(data,bin_edges):
    newList = []
    for num in data:
        i = 0;
        while i < len(bin_edges) and num >= bin_edges[i]:
            i+=1
        newNum = bin_edges[max(0,i-1)]
        newList.append(newNum)
    return newList
```
